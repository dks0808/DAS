#### 딕셔너리가 데이터 프레임에 들어가면 키가 컬럼 이름이 되는 듯

```python
left = pd.DataFrame({"key" : ["foo","foo"], "lval":[1,2]})
right = pd.DataFrame({"key" : ["foo","foo"], "lval":[3,4]})
```

```python
left
```

```python
right
```

```python
merged = pd.merge(left = left, right = right , on = "key") # sql 로 따지면 조인?
```

```python
merged
```

```python
left2 = pd.DataFrame({"key" : ["foo","bar"], "lval":[3,4]})
right2 = pd.DataFrame({"key" : ["foo","bar"], "rval":[5,6]})
```

#### Group by 관련 내용

```python
df5 = pd.DataFrame({"A" : ["foo","bar","foo","bar",
                           "foo","bar","foo","foo"]*2,
                    "B" : ["A","B","C","D"]*4,
                    "C" : ["one" ,"one", "two", "three",
                           "two", "two", "one" , "three"]*2,
                    "D" : np.random.randn(16),
                    "E" : np.random.randn(16),})
```

```python
df5
```

```python
df5.groupby(["A","B","C"]).mean()
```

```python
df5.groupby(["A","B"]).min()
```

```python
df5.groupby(["A","B"]).agg(["min", "max"])
```

```python
pd.pivot_table(df5, values = "D" , index = ["A","B"], columns = ["C"]) 
```

```python
pd.pivot_table(df5, values = "D" , index = "A" , columns = "C", aggfunc = "sum") # 꼭 리스트가 아니어도 되는 듯 리스트도 되고
```
