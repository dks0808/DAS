# 파이썬 포맷팅 방법 

- 파이썬에서는 문자열을 출력하기 위해서 다양한 포맷팅 방법이 있습니다.
1. %를 활용한 포맷팅
2. format()을 활용한 포맷팅
3. f-string 을 활용한 포맷팅

### 1. %를 활용한 포맷팅

- 문자열 뒤에 % 기호를 붙이고 값을 붙이면 됩니다.
    - %s : 문자열
    - %d : 정수
    - %f : 부동소수점 실수

```python
"파이썬 포맷팅 %s" % "테스트" 
```

```python
"1 2 3 4 5 %d" % 6
```

```python
"1/3 = %f" % 0.3333
```

```python
"전체 테스트 문자열 : %s / 정수형 : %d / 실수형 : %f" %("문자열", 123, 0.111)
```

### 2. format()을 활용한 포맷팅
- 앞서 나온 % 기호 대신 {}를 사용하며 따로 형식을 지정하지 않으므로 편하게 사용가능합니다.

```python
"파이썬 포맷팅 {}".format("테스트")
```

```python
"1 2 3 4 5 {}".format(6)
```

```python
"1/3 = {}".format(0.3333)
```

```python
"전체 테스트 문자열 : {} / 정수형 : {} / 실수형 : {}".format("문자열", 123, 0.111)
```

### 3. f-string 을 활용한 포맷팅

- 문자열 시작전 f를 붙여주고 변수를 활용합니다.



```python
var1 = '테스트'
print(f"파이썬 포맷팅 {var1}")
```

```python
var2 = 6
print("1 2 3 4 5 {var2}")
```

```python
var3 = 0.333
print(f"1/3 = {var3}")
```

```python
var4, var5, var6 = "문자열", 123, 0.111
print(f"전체 테스트 문자열 : {var4} / 정수형 : {var5} / 실수형 : {var6}")
```
