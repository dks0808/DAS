# 박스 플랏이란?
기술 통계학에서 '상자 수염 그림' 또는 '상자 그림'은 수치적 자료를 표현하는 그래프입니다. 이 그래프는 데이터로부터 얻어낸 통계량를 활용하고 있습니다.
1. 최솟값
2. 제 1사분위
3. 제 2사분위
4. 제 3사분위
5. 최댓값

위의 통계량은 참고 링크(https://velog.io/@suminwooo/%ED%8C%8C%EC%9D%B4%EC%8D%AC-Pandas-%ED%99%9C%EC%9A%A9)에서 분위수에서 확인할 수 있습니다.

- 여기서 IQR은 데이터의 중간 50% (Q3 - Q1)를 의미합니다.
- 따라서 값들을 통해 이상치가 되는 부분을 설정할 수 있습니다.
- 아래의 이미지에서 Minimum, Maximum은 실제로 최소값 최대값을 의미하지 않습니다. 이 값들은 IQR을 통해 생성된 값입니다.
- 실제 최솟값, 최대값은 초록점처럼 이상치가 될수도 있고 범위 내에 포함될 수 있습니다.
